```python
import pandas as pd
import seaborn as sms
import numpy as np
from sklearn.preprocessing import FunctionTransformer
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan-train.csv")
dataset.head(3)
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset['CoapplicantIncome_ft'] = pd.DataFrame(ft.transform(dataset[["CoapplicantIncome"]]))
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Before")
sms.distplot(dataset["CoapplicantIncome"])

plt.subplot(1, 2, 2)
plt.title("After")
sms.distplot(dataset["CoapplicantIncome_ft"])

plt.show()
```
