```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import seaborn as sms
import matplotlib.pyplot as plt
```

# Standardization
### z = (x-mean)/StandardDeviation

```python
dataset = pd.read_csv("loan-train.csv")
dataset.head(3)
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mode()[0], inplace=True)
```

```python
plt.figure(figsize=(10, 5))
sms.distplot(dataset["ApplicantIncome"])
plt.plot()
```

```python
sc = StandardScaler()
sc.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = sc.transform(dataset[["ApplicantIncome"]])
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Before")
sms.distplot(dataset["ApplicantIncome"])
plt.plot()

plt.subplot(1, 2, 2)
plt.title("After")
sms.distplot(dataset["ApplicantIncome_ss"])
plt.plot()
```

# Min-Max Scaling
### (x-min)/max-min

```python
dataset = pd.read_csv("loan-train.csv")
dataset.head(3)
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mode()[0], inplace=True)
```

```python
mms = MinMaxScaler()
mms.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_mm"] =  pd.DataFrame(mms.transform(dataset[["ApplicantIncome"]]))
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Before")
sms.distplot(dataset["ApplicantIncome"])
plt.plot()

plt.subplot(1, 2, 2)
plt.title("After")
sms.distplot(dataset["ApplicantIncome_mm"])
plt.plot()
```
