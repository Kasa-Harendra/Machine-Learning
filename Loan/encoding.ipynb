```python
import pandas as pd
```

```python
dataset = pd.read_csv("loan-train.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0], inplace=True)
dataset["Married"].fillna(dataset["Married"].mode()[0], inplace=True)
```

## get_dummies

```python
en_data = dataset[["Gender"]]
ar = pd.get_dummies(en_data)
ar.head()
```

## OneHotEncoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
dataset['Gender-coded'] = ar
dataset.head()
```

## LabelEncoder

```python
dataset["Property_Area"].fillna(dataset["Property_Area"].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
en_data = dataset[["Property_Area"]]
le.fit(en_data)
```

```python
ar = le.transform(en_data)
ar
```

```python
dataset["Property_Area_encoded"] = ar
dataset.head()
```

## Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[["Graduate", "Not Graduate"]])
en_data = dataset[["Education"]]
oe.fit(en_data)
```

```python
dataset["Edu_sklearn"] = oe.transform(en_data)
dataset.head()
```

## Map

```python
ord_data = {'Graduate':1, 'Not Graduate':0}
dataset["Edu_map"] = dataset["Education"].map(ord_data)
dataset.head()
```
