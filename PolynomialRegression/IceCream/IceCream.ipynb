```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("IceCream.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="Temperature (°C)", y="Ice Cream Sales (units)", data=dataset)
plt.show()
```

```python
dataset.corr()
```

```python
x = dataset[["Temperature (°C)"]]
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)
```

```python
x.shape
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

y = dataset["Ice Cream Sales (units)"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
plt.scatter(dataset["Temperature (°C)"], dataset["Ice Cream Sales (units)"])
plt.plot(dataset["Temperature (°C)"], lr.predict(x), c="r")
plt.show()
```
