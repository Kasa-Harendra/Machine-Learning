```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("train.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(x="x", data=dataset)
plt.subplot(1, 2, 2)
sns.boxplot(x="y", data=dataset)
plt.show()
```

```python
IQR = 75.000000 - 25.000000
min_range = 25 - 1.5*IQR
max_range = 75 + 1.5*IQR
min_range, max_range
```

```python
dataset = dataset[(dataset["x"]>=min_range) & (dataset["x"]<=max_range)]
```

```python
x = dataset[['x']]
y = dataset[['y']]
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(x="x", data=dataset)
plt.subplot(1, 2, 2)
sns.boxplot(x="y", data=dataset)
plt.show()
```

```python
sns.scatterplot(x="x", y="y", data=dataset)
plt.show()
```
